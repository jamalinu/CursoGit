```python
import pandas as pd

# Creamos un diccionario con tus frases
# Como eres experto, he añadido la columna 'Dialecto' para diferenciar
datos_proyecto = {
    "Tamazight": ["Azul", "Amek tetat?", "Manwen isem nnek?", "Ar tufat"],
    "Dialecto": ["Estándar", "Taqbaylit", "Tarifit", "Estándar"],
    "English": ["Hello", "How are you?", "What is your name?", "See you later"],
    "Arabic": ["مرحباً", "كيف حالك؟", "ما اسمك؟", "إلى اللقاء"],
    "Català": ["Hola", "Com estàs?", "Com et dius?", "Aravé"],
    "Español": ["Hola", "¿Cómo estás?", "¿Cómo te llamas?", "Hasta luego"],
    "Français": ["Salut", "Comment ça va?", "Comment tu t'appelles?", "À plus tard"],
    "Categoria": ["Saludo", "Pregunta", "Pregunta", "Despedida"]
}

# Convertimos este diccionario en una Tabla (DataFrame)
df = pd.DataFrame(datos_proyecto)

# Mostramos la tabla en pantalla
print("¡Felicidades! Este es tu primer Dataset para Google:")
df
```

```python
# 1. Calculamos la longitud de cada frase en Tamazight
df['Longitud'] = df['Tamazight'].apply(len)

# 2. Importamos una herramienta para hacer dibujos (gráficos)
import matplotlib.pyplot as plt

# 3. Creamos un gráfico de barras
plt.bar(df['Tamazight'], df['Longitud'], color='skyblue')

# 4. Ponemos etiquetas bonitas
plt.title('Longitud de las frases en Tamazight')
plt.xlabel('Frases')
plt.ylabel('Número de letras')

# 5. Mostramos el gráfico
plt.show()
```

```python
# 1. Creamos las nuevas frases que nos has dado
nuevas_frases = [
    {"Tamazight": "Manis kkin?", "Dialecto": "Tarifit", "English": "Where did they go?", "Español": "¿Por dónde pasaron / dónde están?", "Categoria": "Pregunta"},
    {"Tamazight": "Wassi mermi id ghayasen", "Dialecto": "Tarifit", "English": "It is unknown when they will come", "Español": "No se sabe cuándo vendrán", "Categoria": "Declarativa"},
    {"Tamazight": "Mri dyusa iri zrighth", "Dialecto": "Tarifit", "English": "If he had come, I would have seen him", "Español": "Si hubiera venido, lo habría visto", "Categoria": "Condicional"}
]

# 2. Las añadimos a nuestra tabla principal
df_nuevas = pd.DataFrame(nuevas_frases)
df = pd.concat([df, df_nuevas], ignore_index=True)

# 3. Función para contar palabras (no solo letras)
df['Num_Palabras'] = df['Tamazight'].apply(lambda x: len(x.split()))

print("¡Dataset profesional ampliado!")
df[['Tamazight', 'Dialecto', 'Categoria', 'Num_Palabras']]
```

```python
from google.colab import files

# Guardamos el DataFrame final en un archivo CSV
df.to_csv('tamazight_multilingual_bridge.csv', index=False)

# Lo descargamos a tu carpeta de Descargas
files.download('tamazight_multilingual_bridge.csv')
print("¡Archivo descargado! Ya tienes la base de tu proyecto.")
```

```python
def buscador_multilingue(palabra_buscada):
    # Buscamos la palabra en la columna de Español
    # .str.contains ayuda a encontrarla aunque no esté escrita exacta
    resultado = df[df['Español'].str.contains(palabra_buscada, case=False)]

    if not resultado.empty:
        print(f"--- Resultados para: {palabra_buscada} ---")
        for index, fila in resultado.iterrows():
            print(f"Tamazight ({fila['Dialecto']}): {fila['Tamazight']}")
            print(f"Català: {fila['Català']}")
            print(f"English: {fila['English']}")
            print(f"Comentario: {fila['Comentario_Lingüístico'] if 'Comentario_Lingüístico' in df.columns else 'N/A'}")
            print("-" * 30)
    else:
        print("Lo siento, esa palabra no está en nuestro puente lingüístico todavía.")

# ¡Pruébalo! Escribe una palabra que esté en tu tabla
buscador_multilingue("llam")
```
