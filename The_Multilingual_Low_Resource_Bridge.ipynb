```python
from collections import Counter
import re # Import the regular expression module for cleaning text
import matplotlib.pyplot as plt
import seaborn as sns

# 0. Limpiamos las frases de Tamazight para un mejor análisis de frecuencia
# Eliminamos signos de puntuación y convertimos a minúsculas
df['Tamazight_Limpio'] = df['Tamazight'].apply(lambda x: re.sub(r'[.,!?\-]', '', x).lower())

# 1. Juntamos todas las frases de Tamazight limpias en un solo texto largo
texto_completo = " ".join(df['Tamazight_Limpio'])

# 2. Rompemos el texto en palabras individuales
palabras = texto_completo.split()

# 3. Contamos cuántas veces sale cada palabra
frecuencia = Counter(palabras)

# 4. Mostramos las 5 palabras más comunes
print("📊 Análisis de Frecuencia en Tamazight:")
for palabra, conteo in frecuencia.most_common(5):
    print(f"La palabra '{palabra}' aparece {conteo} veces.")

# 5. Guardamos este análisis en un pequeño gráfico para GitHub
palabras_top = [p[0] for p in frecuencia.most_common(10)]
conteos_top = [p[1] for p in frecuencia.most_common(10)]

plt.figure(figsize=(10,5))
sns.barplot(x=conteos_top, y=palabras_top, palette="viridis")
plt.title("Top 10 Palabras más frecuentes en nuestro Corpus de Tamazight")
plt.xlabel("Frecuencia")
plt.ylabel("Palabra")
plt.show()
```

```python
# 1. Creamos un "diccionario de conocimientos" basado en tu experiencia nativa
conocimiento_dialectos = {
    "Tarifit": ["mamch", "mermi", "ghayasen", "mri", "zrighth", "manis"],
    "Taqbaylit": ["amek", "tetat", "azul"]
}

def detectar_dialecto(frase):
    frase = frase.lower()
    # Miramos palabra por palabra si coincide con nuestro conocimiento
    for dialecto, palabras_clave in conocimiento_dialectos.items():
        for palabra in palabras_clave:
            if palabra in frase:
                return dialecto
    return "Desconocido o Estándar"

# 2. Probamos nuestro detector con una frase nueva
frase_prueba = "Mermi id ghayasen?"
resultado = detectar_dialecto(frase_prueba)

print(f"La frase: '{frase_prueba}'")
print(f"🤖 El ordenador dice que el dialecto es: {resultado}")
```

```python
from textblob import TextBlob

def analizar_sentimiento(texto_ingles):
    analisis = TextBlob(texto_ingles)
    # Polarity va de -1 (muy triste/malo) a 1 (muy feliz/bueno)
    if analisis.sentiment.polarity > 0:
        return "Positivo 😊"
    elif analisis.sentiment.polarity < 0:
        return "Negativo 😟"
    else:
        return "Neutral 😐"

# Lo aplicamos a nuestro dataset usando la traducción al inglés como referencia
df['Sentimiento'] = df['English'].apply(analizar_sentimiento)

print("Análisis de Sentimiento completado:")
df[['Tamazight', 'English', 'Sentimiento']].tail(5)
```

```python
# Definimos tus Stopwords (Añade tú las que consideres como experto)
stopwords_tamazight = ["d", "n", "nney", "i", "a"]

# Función para limpiar el análisis de frecuencia
def filtrar_palabras_clave(texto):
    # Manejamos los valores NaN o no-string que causan el error
    if pd.isna(texto) or not isinstance(texto, str):
        return ""

    palabras = texto.lower().split()
    # Solo nos quedamos con palabras que NO estén en la lista de stopwords
    filtradas = [p for p in palabras if p not in stopwords_tamazight]
    return " ".join(filtradas)

# Es importante re-calcular Tamazight_Limpio después de cualquier adición de filas
# para asegurar que todas las entradas son strings y están limpias.
# Asegúrate de importar pandas si no lo has hecho en el notebook.
import pandas as pd
import re # Aseguramos que re está importado para re.sub

df['Tamazight_Limpio'] = df['Tamazight'].astype(str).apply(lambda x: re.sub(r'[.,!?\-]', '', x).lower())

df['Tamazight_Key_Words'] = df['Tamazight_Limpio'].apply(filtrar_palabras_clave)

print("✅ Ahora tienes una columna con las palabras con más 'carga de significado'.")
df[['Tamazight', 'Tamazight_Key_Words']].head()
```

```python
# 1. Creamos el diccionario de equivalencias (Mapeo)
tifinagh_map = {
    'a': 'ⴰ', 'b': 'ⴱ', 'g': 'ⴳ', 'd': 'ⴷ', 'e': 'ⴻ', 'f': 'ⴼ',
    'k': 'ⴽ', 'h': 'ⵀ', 'i': 'ⵉ', 'j': 'ⵊ', 'l': 'ⵍ', 'm': 'ⵎ',
    'n': 'ⵏ', 'u': 'ⵓ', 'r': 'ⵔ', 's': 'ⵙ', 't': 'ⵜ', 'q': 'ⵇ',
    'w': 'ⵡ', 'x': 'ⵅ', 'y': 'ⵢ', 'z': 'ⵣ', ' ': ' '
}

def a_tifinagh(texto):
    texto = texto.lower()
    tifinagh_result = ""
    for letra in texto:
        # Si la letra está en nuestro mapa, la cambiamos; si no, la dejamos igual
        tifinagh_result += tifinagh_map.get(letra, letra)
    return tifinagh_result

# 2. Aplicamos la conversión a todo nuestro Dataset
df['Tifinagh'] = df['Tamazight'].apply(a_tifinagh)

print("✨ ¡Dataset actualizado con escritura original!")
df[['Tamazight', 'Tifinagh', 'Dialecto']]
```
